```python
import pandas #to read the datafame
```

```python
anz = pandas.read_csv('../data/gapminder_gdp_oceania.csv', index_col='country')
europe = pandas.read_csv('../data/gapminder_gdp_europe.csv', index_col='country')
```

```python
#create plots in the same cell
%matplotlib inline 
import matplotlib.pyplot as plt
```

```python
anz.T.plot()
europe.T.plot()
```

```python
anz.columns
europe.columns
```

```python
# format column names to only the number
years=europe.columns.str.strip('gdpPercap_')
europe.columns=years
europe.columns
```

```python
anz.T.plot()
```

```python
plt.style.available
```

```python
plt.style.use('ggplot')
europe.plot()
europe.plot().set_color_cycle(['red', 'black', 'yellow'])
```

```python
plt.style.use('ggplot')
europe.plot(kind='box')
europe.plot(kind='bar')
europe.plot(kind='line')
```

```python
somecountyr=europe.loc[('Italy','Belgium','Ireland','Iceland'),]
```

```python
plt.plot(years,somecountyr.iloc[0],'g--',label='Italy')
plt.plot(years, somecountyr.iloc[1], 'y--', label='Ireland')
plt.plot(years, somecountyr.iloc[2], 'r--', label='Iceland')
plt.legend(loc= 'best')
plt.xlabel('Years')
plt.ylabel('GDP per Cap')
plt.title('GDP per cap in Italy, Ireland and Iceland')
plt.savefig('../plot/GDP_i_countries.png')
```
